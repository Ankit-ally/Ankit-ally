### Reading PDF Content  ### only uses for pdf reading long pdf in very short crisp and clear for understanding

To interact with PDF files, we'll use the `PyPDF2` library. This library allows us to extract text, pages, and other information from PDF documents. Make sure the PDF file `NEP_Final_English_0.pdf` is uploaded to your Colab environment or is accessible at the specified path.

````python
import PyPDF2
import os

pdf_file_path = '/NEP_Final_English_0.pdf'

# Check if the file exists
if not os.path.exists(pdf_file_path):
    print(f"Error: The file '{pdf_file_path}' does not exist. Please upload it to your Colab environment or ensure the path is correct.")
else:
    try:
        # Open the PDF file in binary read mode
        with open(pdf_file_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)

            # Print general info about the PDF
            print(f"Number of pages: {len(reader.pages)}")

            # Extract text from the first page as an example
            if reader.pages:
                first_page = reader.pages[0]
                text = first_page.extract_text()
                print("\n--- Extracted Text from Page 1 ---")
                print(text[:1000]) # Print first 1000 characters of text
                print("\n--- End of Page 1 Extract ---")
            else:
                print("The PDF contains no pages.")
    except Exception as e:
        print(f"An error occurred while reading the PDF: {e}")
````

### Extracting All Text from the PDF

To perform a comprehensive analysis, we need to extract the complete text content from all pages of the PDF. This will allow us to process the entire document for keywords, themes, or to generate summaries.

````python
full_text = []

if os.path.exists(pdf_file_path):
    try:
        with open(pdf_file_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            for page_num in range(len(reader.pages)):
                page = reader.pages[page_num]
                full_text.append(page.extract_text())

        # Join all extracted text into a single string
        document_text = "\n".join(full_text)

        print(f"Successfully extracted text from {len(reader.pages)} pages.")
        print("\n--- First 500 characters of the full document text ---")
        print(document_text[:500])
        print("\n--- Last 500 characters of the full document text ---")
        print(document_text[-500:])

    except Exception as e:
        print(f"An error occurred while extracting all text from the PDF: {e}")
else:
    print(f"Error: The file '{pdf_file_path}' does not exist. Cannot extract full text.")
````

````python
# Extraction
import pdfplumber

def extract_text(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() or ""
    return text

# Using the previously confirmed existing PDF file for extraction with pdfplumber
text = extract_text(pdf_file_path) # pdf_file_path refers to '/NEP_Final_English_0.pdf'
print(text[:1000])
````

````python
## text preprocessing
import re
# RecursiveCharacterTextSplitter is now imported globally in BVh6EpeG9Bgn

def clean_text(text):
    text = re.sub(r'\n+', '\n', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

def split_text(text):
    # This relies on RecursiveCharacterTextSplitter being globally imported
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=500,
        chunk_overlap=100
    )
    chunks = splitter.split_text(text)
    return chunks

cleaned_text = clean_text(text)

def classify_topic(text):
    prompt = f"""
    Classify this education regulation into one category:
    [Accreditation, Scholarship, Curriculum, Faculty Policy, Examination, Admissions]

    Text:
    {text[:2000]}
    """
    response = model.generate_content(prompt)
    return response.text

def generate_summary(text):
    prompt = f"""
    Simplify this regulation into 10-15 lines.
    Make it easy for students and faculty to understand.

    Text:
    {text[:3000]}
    """
    response = model.generate_content(prompt)
    return response.text

def stakeholder_analysis(text):
    prompt = f"""
    Analyze stakeholders impacted by this regulation:

    Identify impact on:
    - Students
    - Faculty
    - Institutions
    - Administrators

    Text:
    {text[:2000]}
    """
    response = model.generate_content(prompt)
    return response.text

def impact_forecast(text):
    prompt = f"""
    Provide impact analysis:
    - Short-term (0-1 year)
    - Medium-term (1-5 years)
    - Long-term (>5 years)

    Text:
    {text[:2000]}
    """
    response = model.generate_content(prompt)
    return response.text
````

````python
from textblob import TextBlob

def analyze_sentiment(text):
    blob = TextBlob(text)
    # Polarity is a float within the range [-1.0, 1.0] where -1.0 is negative and 1.0 is positive.
    # Subjectivity is a float within the range [0.0, 1.0] where 0.0 is very objective and 1.0 is very subjective.
    return blob.sentiment.polarity, blob.sentiment.subjectivity

# Perform sentiment analysis on the first document
polarity_doc1, subjectivity_doc1 = analyze_sentiment(cleaned_text)
print(f"Sentiment for '/NEP_Final_English_0.pdf':")
print(f"  Polarity: {polarity_doc1:.2f}")
print(f"  Subjectivity: {subjectivity_doc1:.2f}")

# Perform sentiment analysis on the second document
polarity_doc2, subjectivity_doc2 = analyze_sentiment(new_document_text)
print(f"\nSentiment for '/JNRID2506023.pdf':")
print(f"  Polarity: {polarity_doc2:.2f}")
print(f"  Subjectivity: {subjectivity_doc2:.2f}")
````

### Document Similarity Heatmap

````python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt

# Create a list of documents for vectorization
documents_for_similarity = [cleaned_text, new_document_text]
document_names = ['NEP_Final_English_0.pdf', 'JNRID2506023.pdf']

# Initialize TF-IDF Vectorizer
# min_df ignores terms that appear in less than 2 documents (or a given fraction)
# stop_words='english' removes common English stop words
vectorizer = TfidfVectorizer(stop_words='english', min_df=1)

# Fit and transform the documents
tfidf_matrix = vectorizer.fit_transform(documents_for_similarity)

# Calculate cosine similarity
cosine_sim_matrix = cosine_similarity(tfidf_matrix)

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cosine_sim_matrix, annot=True, cmap='viridis',
            xticklabels=document_names, yticklabels=document_names,
            fmt=".2f", linewidths=.5)
plt.title('Document Similarity Heatmap (Cosine Similarity)')
plt.show()
````

````python
def export_pipeline_results_to_file(document_text_to_export, output_filename):
    """
    Exports the pipeline analysis results for a given document text to a text file.
    """
    with open(output_filename, 'w', encoding='utf-8') as f:
        f.write(f"--- Pipeline Results for JNRID2506023.pdf ---\n\n")

        # Summary
        summary = generate_summary(document_text_to_export)
        f.write("📌 Summary:\n")
        f.write(summary)
        f.write("\n\n")

        # Topic
        topic = classify_topic(document_text_to_export)
        f.write("📌 Topic:\n")
        f.write(topic)
        f.write("\n\n")

        # Stakeholders
        stakeholders = stakeholder_analysis(document_text_to_export)
        f.write("📌 Stakeholders:\n")
        f.write(stakeholders)
        f.write("\n\n")

        # Impact Forecast
        impact = impact_forecast(document_text_to_export)
        f.write("📌 Impact:\n")
        f.write(impact)
        f.write("\n\n")
    print(f"Pipeline results exported to '{output_filename}'")
````

````python
# similiraty search Retrival
def retrieve_context(query, k=3):
    docs = vector_db.similarity_search(query, k=k)
    context = "\n".join([doc.page_content for doc in docs])
    return context
````

````python
## RAG + Gemini (CORE MAGIC)
def rag_query(query):
    context = retrieve_context(query)

    prompt = f"""
    Use the following context to answer the question.

    Context:
    {context}

    Question:
    {query}
    """

    response = model.generate_content(prompt)
    return response.text
````

````python
import pdfplumber
import re
from langchain_text_splitters import RecursiveCharacterTextSplitter

def extract_text(pdf_path):
    text = ""
    try:
        with pdfplumber.open(pdf_path) as pdf:
            for page in pdf.pages:
                text += page.extract_text() or ""
    except Exception as e:
        print(f"Error extracting text from {pdf_path}: {e}")
        return ""
    return text

def clean_text(text):
    text = re.sub(r'\n+', '\n', text) # Replace multiple newlines with a single one
    text = re.sub(r'\s+', ' ', text)  # Replace multiple whitespaces with a single one
    return text.strip()

def split_text(text):
    # Initialize RecursiveCharacterTextSplitter with specified chunk size and overlap
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=500,
        chunk_overlap=100
    )
    chunks = splitter.split_text(text)
    return chunks

print("✅ `extract_text`, `clean_text`, and `split_text` functions redefined.")
````

````python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt

# Create a list of documents for vectorization
documents_for_similarity = [cleaned_text, new_document_text]
document_names = ['NEP_Final_English_0.pdf', 'JNRID2506023.pdf']

# Initialize TF-IDF Vectorizer
# min_df ignores terms that appear in less than 1 document (or a given fraction) as we only have 2 documents
# stop_words='english' removes common English stop words
vectorizer = TfidfVectorizer(stop_words='english', min_df=1)

# Fit and transform the documents
tfidf_matrix = vectorizer.fit_transform(documents_for_similarity)

# Calculate cosine similarity
cosine_sim_matrix = cosine_similarity(tfidf_matrix)

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cosine_sim_matrix, annot=True, cmap='viridis',
            xticklabels=document_names, yticklabels=document_names,
            fmt=".2f", linewidths=.5)
plt.title('Document Similarity Heatmap (Cosine Similarity)')
plt.show()
print("✅ Document similarity calculated and visualized.")
````

````python
%%writefile app.py
import streamlit as st
import pdfplumber
import re
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
import google.generativeai as genai

# ── Page Config ───────────────────────────────
st.set_page_config(page_title="ERIA - Regulation Analyzer", layout="wide")

st.title("📊 Education Regulation Impact Analyzer (ERIA)")
st.write("Upload a regulation PDF and chat with it using AI")

# ── Gemini Setup ─────────────────────────────
# Use the already configured GOOGLE_API_KEY from the environment
genai.configure(api_key=GOOGLE_API_KEY) # Use the globally available GOOGLE_API_KEY
model = genai.GenerativeModel("gemini-1.5-flash")

# ── Functions ───────────────────────────────

def extract_text(pdf_file):
    text = ""
    with pdfplumber.open(pdf_file) as pdf:
        for page in pdf.pages:
            text += page.extract_text() or ""
    return text

def clean_text(text):
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

def create_vector_db(text):
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=500,
        chunk_overlap=100
    )
    chunks = splitter.split_text(text)

    embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
    db = FAISS.from_texts(chunks, embeddings)

    return db

def ask_question(db, query):
    docs = db.similarity_search(query, k=3)
    context = "\n".join([doc.page_content for doc in docs])

    prompt = f"""
    You are an education policy expert.

    Context:
    {context}

    Question:
    {query}

    Give clear, structured, simple answer.
    """

    response = model.generate_content(prompt)
    return response.text

# ── Session State ───────────────────────────
if "chat_history" not in st.session_state:
    st.session_state.chat_history = []

if "db" not in st.session_state:
    st.session_state.db = None

# ── File Upload ─────────────────────────────
uploaded_file = st.file_uploader("📥 Upload PDF", type="pdf")

if uploaded_file:
    with st.spinner("Processing document..."):
        text = extract_text(uploaded_file)
        text = clean_text(text)

        db = create_vector_db(text)
        st.session_state.db = db

    st.success("✅ Document processed successfully!")

# ── Chat UI ────────────────────────────────
st.subheader("💬 Chat with Regulation")

user_query = st.text_input("Ask a question:")

if st.button("Ask"):
    if st.session_state.db is None:
        st.warning("Please upload a PDF first")
    else:
        answer = ask_question(st.session_state.db, user_query)

        st.session_state.chat_history.append(("You", user_query))
        st.session_state.chat_history.append(("AI", answer))

# ── Display Chat ───────────────────────────
for role, msg in st.session_state.chat_history:
    if role == "You":
        st.markdown(f"**🧑 You:** {msg}")
    else:
        st.markdown(f"**🤖 AI:** {msg}")
````

## Running Streamlit Applications in Colab

Streamlit applications are essentially Python scripts that leverage the Streamlit library to create interactive web UIs directly from your Python code. Unlike traditional web development, you don't need to write HTML, CSS, or JavaScript.

### Standard Execution

Typically, to run a Streamlit application from your local terminal, you would execute a command like this:

```bash
streamlit run your_app_name.py
```

Where `your_app_name.py` is the Python file containing your Streamlit code. This command starts a local web server, and your browser automatically opens to display the application, usually at `http://localhost:8501`.

### Running in Google Colab

Running Streamlit applications within a Google Colab environment has a few specific considerations:

1.  **Save Code to a `.py` file**: Colab cells are executed sequentially. To run a Streamlit app, all its code needs to be contained within a single Python script file. Therefore, you first need to save your Streamlit application's Python code (e.g., from an earlier Colab cell or direct input) into a `.py` file (e.g., `app.py`) within the Colab environment.

2.  **Execute the Script**: Once the code is saved, you run it using a shell command within a Colab code cell, prefixing it with an exclamation mark `!`:

    ```python
    !streamlit run app.py
    ```

    This command starts the Streamlit server inside the Colab runtime.

3.  **External Access with `ngrok`**: By default, the Streamlit app runs on a port (usually 8501) within the Colab runtime, which isn't directly accessible from your local web browser. To expose this local server to the public internet, a tunneling service like `ngrok` is commonly used. `ngrok` creates a secure tunnel from a public endpoint to a locally running web service. When `ngrok` is used, it provides a public URL (e.g., `https://xxxx-xxxx-xxxx.ngrok-free.app`) that you can access from anywhere to view your Streamlit application.

    Often, the `streamlit run` command can be combined or followed by commands that set up `ngrok` or similar services, making the URL generation semi-automatic.

### Where the UI Appears

After successfully running the `streamlit run` command (and potentially setting up `ngrok`):

*   **Within Colab**: The Streamlit user interface might appear directly within an output iframe below the code cell in your Colab notebook.
*   **Public URL**: If a tunneling service like `ngrok` is used, a public URL will be generated and displayed in the cell's output. You can then open this URL in any web browser to interact with your Streamlit application.

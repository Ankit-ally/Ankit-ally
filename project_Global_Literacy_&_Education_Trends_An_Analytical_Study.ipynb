## Select Countries and Extract Literacy Data

### Subtask:
Select three specific countries (India, China, Venezuela) and extract their adult literacy rates for the year 2001 and the latest available year from the `df_literacy` DataFrame. For the year 2001, if data is not directly available, find the closest preceding year for each country.


**Reasoning**:
To achieve the subtask, I will filter the `df_literacy` DataFrame for the specified countries, then iterate through each country to find their literacy rates for 2001 (or the closest preceding year) and the latest available year, storing the results in a new DataFrame.



````python
target_countries = ['India', 'China', 'Venezuela']
filtered_literacy_data = df_literacy[df_literacy['entity'].isin(target_countries)]

extracted_literacy_data = []

for country in target_countries:
    country_data = filtered_literacy_data[filtered_literacy_data['entity'] == country]

    # For 2001 (or closest preceding year)
    data_2001_or_earlier = country_data[country_data['year'] <= 2001]
    literacy_2001 = None
    year_2001 = None
    if not data_2001_or_earlier.empty:
        closest_2001_data = data_2001_or_earlier.loc[data_2001_or_earlier['year'].idxmax()]
        literacy_2001 = closest_2001_data['Literacy_Rate']
        year_2001 = closest_2001_data['year']
    else:
        print(f"No literacy data found for {country} in or before 2001.")

    # For latest available year
    latest_data = country_data.loc[country_data['year'].idxmax()]
    latest_literacy = latest_data['Literacy_Rate']
    latest_year = latest_data['year']

    extracted_literacy_data.append({
        'Country': country,
        'Year_2001_Approx': year_2001,
        'Literacy_2001_Approx': literacy_2001,
        'Latest_Year': latest_year,
        'Latest_Literacy': latest_literacy
    })

df_extracted_literacy = pd.DataFrame(extracted_literacy_data)

print("Extracted Literacy Data:")
df_extracted_literacy
````

## Calculate Literacy Rate Change

### Subtask:
Calculate the difference in literacy rates for each selected country between the latest available year and the year 2001 (or closest preceding year), representing the 'increase' or 'change' over time. This change will be used to determine the proportion for the pie chart.


**Reasoning**:
To calculate the literacy rate change, I will subtract the 'Literacy_2001_Approx' from the 'Latest_Literacy' for each country and store the result in a new column called 'Literacy_Change' in the `df_extracted_literacy` DataFrame. Then, I will display the updated DataFrame.



````python
df_extracted_literacy['Literacy_Change'] = df_extracted_literacy['Latest_Literacy'] - df_extracted_literacy['Literacy_2001_Approx']

print("Updated Extracted Literacy Data with Change:")
df_extracted_literacy
````

## Generate Interactive Pie Chart

### Subtask:
Create an interactive Altair pie chart. Each slice will represent the proportion of the literacy rate change for each country relative to the total change across the three countries. The chart will include tooltips to display the country and its specific literacy rate change upon hovering.


**Reasoning**:
To generate the interactive pie chart as specified, I will first calculate the 'Percentage_Change' for each country, then use Altair to create a base chart, define the pie slices with colors and tooltips, add text labels for percentages, combine these elements, set a title, and make the chart interactive.



````python
import altair as alt

# Calculate 'Percentage_Change'
df_extracted_literacy['Percentage_Change'] = (df_extracted_literacy['Literacy_Change'] / df_extracted_literacy['Literacy_Change'].sum()) * 100

# Create a base chart
base = alt.Chart(df_extracted_literacy).encode(
    theta=alt.Theta("Literacy_Change", stack=True)
)

# Create the pie chart (arcs)
pie = base.mark_arc(outerRadius=120).encode(
    color=alt.Color("Country:N"),
    order=alt.Order("Literacy_Change", sort="descending"),
    tooltip=["Country:N", "Literacy_Change:Q", alt.Tooltip("Percentage_Change:Q", format=".2f")])

# Add text labels
text = base.mark_text(radius=140).encode(
    text=alt.Text("Percentage_Change:Q", format=".1f"),
    order=alt.Order("Literacy_Change", sort="descending"),
    color=alt.value("black")  # Set the color of the labels to black
)

# Combine the pie chart and the text labels, add title and interactivity
chart_pie = (pie + text).properties(
    title='Proportion of Literacy Rate Change by Country'
).interactive()

chart_pie
````

## Generate Interactive World Map of Illiteracy Rates

### Subtask:
Create an interactive world map using Altair that visualizes the latest illiteracy rates for each country, allowing for hovering to display country names and their respective illiteracy rates, and supporting interactive features like zooming and panning. This step will correct the `SchemaValidationError` by passing the GeoJSON URL directly to `alt.topo_feature`.


**Reasoning**:
To generate the interactive world map of illiteracy rates, I will use Altair as instructed. The previous attempt resulted in a `SchemaValidationError` because `alt.topo_feature` expects a URL or a string representation of the GeoJSON data, not a parsed Python dictionary. I will pass the `geojSON_url` directly to `alt.topo_feature` to resolve this issue and follow the instructions to build the base map, define the color scale, create the choropleth layer with tooltips and a lookup transform, and combine both layers into an interactive chart.



**Reasoning**:
The previous error `SchemaValidationError: 'quantitative' is an invalid value for type` occurred because `type='quantitative'` is not a valid scale type in Altair. For a quantitative scale with a 'heatmap' range, `type='sequential'` is the correct and valid option. I will correct this in the `alt.Scale` definition.

## Calculate Literacy Gender Gap

## Calculate GDP per Schooling Year

### Subtask:
Merge `df_gdp_schooling_cleaned` and `df_adult_youth_literacy_cleaned` on common columns. Then, calculate 'GDP_per_Schooling_Year' by dividing 'GDP_per_Capita' by 'Avg_Years_Schooling' in the merged DataFrame, indicating economic output per year of education.


**Reasoning**:
To calculate the 'GDP_per_Schooling_Year', I will first merge `df_gdp_schooling_cleaned` and `df_adult_youth_literacy_cleaned` using an outer join on 'entity', 'code', and 'year'. Then, I will create the new column by dividing 'GDP_per_Capita' by 'Avg_Years_Schooling' in the resulting DataFrame.



````python
df_gdp_literacy_merged = pd.merge(df_gdp_schooling_cleaned,
                                df_adult_youth_literacy_cleaned,
                                on=['entity', 'code', 'year'],
                                how='outer',
                                suffixes=('_gdp_schooling', '_adult_youth'))

df_gdp_literacy_merged['GDP_per_Schooling_Year'] = df_gdp_literacy_merged['GDP_per_Capita'] / df_gdp_literacy_merged['Avg_Years_Schooling']

print("Head of df_gdp_literacy_merged with 'GDP_per_Schooling_Year' column:")
df_gdp_literacy_merged.head()
````

**Reasoning**:
To continue with feature engineering, I will calculate the 'Youth_Literacy_Average' by taking the mean of 'Youth_Literacy_Male' and 'Youth_Literacy_Female' in `df_adult_youth_literacy_cleaned`, and then display the head of the DataFrame to verify the new column.



````python
df_adult_youth_literacy_cleaned['Youth_Literacy_Average'] = (df_adult_youth_literacy_cleaned['Youth_Literacy_Male'] + df_adult_youth_literacy_cleaned['Youth_Literacy_Female']) / 2

print("Head of df_adult_youth_literacy_cleaned with new 'Youth_Literacy_Average' column:")
df_adult_youth_literacy_cleaned.head()
````

**Reasoning**:
To calculate the 'Literacy_Growth_Rate', I will group the `df_adult_youth_literacy_cleaned` DataFrame by 'entity', sort by 'year', and then compute the percentage change of 'Literacy_Rate' within each group, assigning the result to a new column, and finally display the head of the DataFrame to verify the new feature.



````python
df_adult_youth_literacy_cleaned['Literacy_Growth_Rate'] = df_adult_youth_literacy_cleaned.groupby('entity')['Literacy_Rate'].pct_change() * 100

print("Head of df_adult_youth_literacy_cleaned with new 'Literacy_Growth_Rate' column:")
df_adult_youth_literacy_cleaned.head()
````

**Reasoning**:
To calculate the 'Education_Index', I will use the `df_gdp_literacy_merged` DataFrame and create the new column based on the literacy rates and average years of schooling as specified in the task, and then display its head.



````python
df_gdp_literacy_merged['Education_Index'] = (df_gdp_literacy_merged['Literacy_Rate'] + df_gdp_literacy_merged['Avg_Years_Schooling'] * 10) / 2

print("Head of df_gdp_literacy_merged with new 'Education_Index' column:")
df_gdp_literacy_merged.head()
````

## Implement SQL Query Executor Page

### Subtask:
Develop the code for a Streamlit page that allows users to input SQL queries, execute them against the SQLite database established earlier, and display the results.


**Reasoning**:
To implement the SQL Query Executor page in Streamlit, I need to provide the complete Python code for the `pages/1_SQL_Query_Executor.py` file. This code will establish its own in-memory SQLite database, populate it with the cleaned DataFrames from the notebook using `st.cache_resource` for efficiency, create a user interface for entering and executing SQL queries, display results in an interactive table, and include error handling and schema information as per the instructions.



````python
import streamlit as st
import sqlite3
import pandas as pd

# IMPORTANT ASSUMPTION FOR THIS GENERATED CODE:
# The dataframes `df_adult_youth_literacy_cleaned`, `df_illiterate_pop_cleaned`,
# and `df_gdp_schooling_features` (in their final processed state from the notebook)
# are assumed to be accessible in the Streamlit app's environment.
# In a real-world Streamlit app, you would either load these from persisted files
# (e.g., CSV, Parquet) or re-execute the data preparation steps within the app,
# typically using st.cache_data or st.cache_resource to prevent re-running on every interaction.

@st.cache_resource
def setup_database():
    """
    Initializes an in-memory SQLite database and populates it with data from
    the cleaned and feature-engineered pandas DataFrames.
    This function is cached to run only once per Streamlit app deployment.
    """
    # Create a new in-memory SQLite connection for this Streamlit session
    conn_streamlit = sqlite3.connect(':memory:')

    # Helper function to create table with composite PK and insert data
    def create_and_populate_table(df, table_name, conn_obj, pk_cols):
        # Get column names and their SQLite dtypes, quoting them explicitly
        column_definitions = []
        for col, dtype in df.dtypes.items():
            sqlite_type = 'TEXT'
            if pd.api.types.is_integer_dtype(dtype):
                sqlite_type = 'INTEGER'
            elif pd.api.types.is_float_dtype(dtype):
                sqlite_type = 'REAL'
            column_definitions.append(f'"{col}" {sqlite_type}')

        # Drop table if it already exists to ensure a clean slate
        conn_obj.execute(f'DROP TABLE IF EXISTS {table_name};')

        # Formulate the CREATE TABLE statement with explicitly quoted column names and composite primary key
        pk_str = ', '.join([f'"{col}"' for col in pk_cols])
        create_table_sql = f"""
        CREATE TABLE {table_name} (
            {', '.join(column_definitions)},
            PRIMARY KEY ({pk_str})
        );
        """
        conn_obj.execute(create_table_sql)
        df.to_sql(table_name, conn_obj, if_exists='append', index=False)
        conn_obj.commit()

    # Access global dataframes from the notebook's kernel state
    # This is how the "database established earlier" (in terms of content) is used.
    # The actual Python objects from the notebook's global scope are assumed to be available.
    global df_adult_youth_literacy_cleaned
    global df_illiterate_pop_cleaned
    global df_gdp_schooling_features

    st.info("Setting up in-memory SQLite database...")
    create_and_populate_table(df_adult_youth_literacy_cleaned, 'literacy_rates', conn_streamlit, ['country', 'year'])
    create_and_populate_table(df_illiterate_pop_cleaned, 'illiteracy_population', conn_streamlit, ['country', 'year'])
    create_and_populate_table(df_gdp_schooling_features, 'gdp_schooling_features', conn_streamlit, ['country', 'year'])
    st.success("Database setup complete!")

    return conn_streamlit

# Setup the database (this runs only once when the app starts)
conn = setup_database()

st.title("SQL Query Executor")

st.markdown("""
This page allows you to execute SQL queries against the in-memory SQLite database.
Available tables: `literacy_rates`, `illiteracy_population`, `gdp_schooling_features`.
""")

# Display schema information for available tables
st.subheader("Table Schemas:")
table_names = ['literacy_rates', 'illiteracy_population', 'gdp_schooling_features']
for table_name in table_names:
    st.write(f"**{table_name}**:")
    try:
        schema_df = pd.read_sql_query(f"PRAGMA table_info({table_name});", conn)
        st.dataframe(schema_df, use_container_width=True)
    except Exception as e:
        st.error(f"Could not retrieve schema for {table_name}: {e}")

# Text area for user input
query_input = st.text_area("Enter your SQL query here:", height=200, value="SELECT * FROM literacy_rates LIMIT 10;")

# Execute button
if st.button("Execute Query"):
    if query_input:
        try:
            # Execute the query using pandas read_sql_query
            result_df = pd.read_sql_query(query_input, conn)
            st.subheader("Query Results:")
            st.dataframe(result_df, use_container_width=True)
            st.success("Query executed successfully!")
        except sqlite3.Error as e:
            st.error(f"SQL Query Error: {e}")
        except Exception as e:
            st.error(f"An unexpected error occurred: {e}")
    else:
        st.warning("Please enter an SQL query to execute.")
````

````python
import streamlit as st
import pandas as pd
import altair as alt
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import io

st.set_page_config(
    page_title="EDA Visualizations",
    page_icon="📈",
)

st.title("Exploratory Data Analysis Visualizations")
st.write("View interactive charts and graphs for key educational and economic indicators.")

# User-Agent for data fetching
headers = {'User-Agent': 'Our World In Data data fetch/1.0'}

@st.cache_data
def load_and_prepare_all_data():
    """
    Loads, cleans, and feature-engineers all necessary dataframes.
    This function is cached to run only once.
    """
    # 1. Load initial datasets
    df_literacy = pd.read_csv("https://ourworldindata.org/grapher/literacy-rate-adults.csv?v=1&csvType=full&useColumnShortNames=true", storage_options={'User-Agent': 'Our World In Data data fetch/1.0'})
    youth_literacy_url = "https://ourworldindata.org/grapher/literacy-rate-of-young-men-and-women.csv?v=1&csvType=full&useColumnShortNames=true"
    response_youth_literacy = requests.get(youth_literacy_url, headers=headers)
    response_youth_literacy.raise_for_status()
    df_youth_literacy = pd.read_csv(io.StringIO(response_youth_literacy.text))

    illiterate_pop_url = "https://ourworldindata.org/grapher/literate-and-illiterate-world-population.csv?v=1&csvType=full&useColumnShortNames=true"
    response_illiterate_pop = requests.get(illiterate_pop_url, headers=headers)
    response_illiterate_pop.raise_for_status()
    df_illiterate_pop = pd.read_csv(io.StringIO(response_illiterate_pop.text))

    gdp_per_capita_url = "https://ourworldindata.org/grapher/gdp-per-capita-worldbank.csv?v=1&csvType=full&useColumnShortNames=true"
    response_gdp = requests.get(gdp_per_capita_url, headers=headers)
    response_gdp.raise_for_status()
    df_gdp = pd.read_csv(io.StringIO(response_gdp.text))

    new_data_url = "https://ourworldindata.org/grapher/literacy-rates-vs-average-years-of-schooling.csv?v=1&csvType=full&useColumnShortNames=true"
    df_literacy_schooling = pd.read_csv(new_data_url, storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})

    # 2. Cleaning and Preprocessing (as done in notebook)

    # df_literacy cleaning
    df_literacy = df_literacy.rename(columns={'adult_literacy_rate__population_15plus_years__both_sexes__pct__lr_ag15t99': 'Literacy_Rate'})

    # df_youth_literacy cleaning
    df_youth_literacy_cleaned = df_youth_literacy.dropna(subset=['youth_literacy_rate__population_15_24_years__male__pct__lr_ag15t24_m', 'youth_literacy_rate__population_15_24_years__female__pct__lr_ag15t24_f']).copy()

    # Merging df_literacy and df_youth_literacy
    df_adult_youth_literacy = pd.merge(df_literacy, df_youth_literacy_cleaned, on=['entity', 'code', 'year'], how='outer', suffixes=('_adult', '_youth'))

    # df_adult_youth_literacy_cleaned (from cleaning stage)
    critical_literacy_cols = [
        'Literacy_Rate',
        'youth_literacy_rate__population_15_24_years__male__pct__lr_ag15t24_m',
        'youth_literacy_rate__population_15_24_years__female__pct__lr_ag15t24_f'
    ]
    df_adult_youth_literacy_cleaned = df_adult_youth_literacy.dropna(subset=critical_literacy_cols).copy()
    df_adult_youth_literacy_cleaned = df_adult_youth_literacy_cleaned.rename(columns={
        'youth_literacy_rate__population_15_24_years__male__pct__lr_ag15t24_m': 'Youth_Literacy_Male',
        'youth_literacy_rate__population_15_24_years__female__pct__lr_ag15t24_f': 'Youth_Literacy_Female'
    })
    df_adult_youth_literacy_cleaned = df_adult_youth_literacy_cleaned[
        (df_adult_youth_literacy_cleaned['year'] >= 1990) &
        (df_adult_youth_literacy_cleaned['year'] <= 2023)
    ].copy()

    # df_illiterate_pop_cleaned (from cleaning stage)
    df_illiterate_pop_cleaned = df_illiterate_pop.dropna(subset=['illiteracy_rate', 'literacy_rate']).copy()
    df_illiterate_pop_cleaned = df_illiterate_pop_cleaned[
        (df_illiterate_pop_cleaned['year'] >= 1990) &
        (df_illiterate_pop_cleaned['year'] <= 2023)
    ].copy()

    # df_gdp_schooling_cleaned (from cleaning stage)
    df_gdp_schooling = pd.merge(
        df_gdp, df_literacy_schooling,
        on=['entity', 'code', 'year'], how='outer',
        suffixes=('_gdp', '_schooling')
    )
    critical_gdp_schooling_cols = [
        'ny_gdp_pcap_pp_kd',
        'mf_youth_and_adults__15_64_years__average_years_of_education'
    ]
    df_gdp_schooling_cleaned = df_gdp_schooling.dropna(subset=critical_gdp_schooling_cols).copy()
    df_gdp_schooling_cleaned = df_gdp_schooling_cleaned.rename(columns={
        'ny_gdp_pcap_pp_kd': 'GDP_per_Capita',
        'mf_youth_and_adults__15_64_years__average_years_of_education': 'Avg_Years_Schooling',
        'population_historical': 'Population_Historical'
    })
    df_gdp_schooling_cleaned = df_gdp_schooling_cleaned[
        (df_gdp_schooling_cleaned['year'] >= 1990) &
        (df_gdp_schooling_cleaned['year'] <= 2023)
    ].copy()

    # 3. Feature Engineering (as done in notebook)
    df_illiterate_pop_cleaned['Illiteracy_Percentage'] = df_illiterate_pop_cleaned['illiteracy_rate']
    df_adult_youth_literacy_cleaned['Literacy_Gender_Gap'] = abs(df_adult_youth_literacy_cleaned['Youth_Literacy_Male'] - df_adult_youth_literacy_cleaned['Youth_Literacy_Female'])
    df_adult_youth_literacy_cleaned['Youth_Literacy_Average'] = (df_adult_youth_literacy_cleaned['Youth_Literacy_Male'] + df_adult_youth_literacy_cleaned['Youth_Literacy_Female']) / 2
    df_adult_youth_literacy_cleaned['Literacy_Growth_Rate'] = df_adult_youth_literacy_cleaned.groupby('entity')['Literacy_Rate'].pct_change() * 100

    # Merge df_gdp_schooling_cleaned and df_adult_youth_literacy_cleaned for combined features
    df_gdp_literacy_merged = pd.merge(
        df_gdp_schooling_cleaned,
        df_adult_youth_literacy_cleaned,
        on=['entity', 'code', 'year'],
        how='outer',
        suffixes=('_gdp_schooling', '_adult_youth')
    )

    df_gdp_literacy_merged['GDP_per_Schooling_Year'] = df_gdp_literacy_merged['GDP_per_Capita'] / df_gdp_literacy_merged['Avg_Years_Schooling']
    # Recalculate Youth_Literacy_Average on df_gdp_literacy_merged as it might have NaNs from outer join
    df_gdp_literacy_merged['Youth_Literacy_Average'] = (df_gdp_literacy_merged['Youth_Literacy_Male'] + df_gdp_literacy_merged['Youth_Literacy_Female']) / 2

    # Ensure Literacy_Rate and Avg_Years_Schooling are not NaN before calculating Education_Index
    df_gdp_literacy_merged['Education_Index'] = (
        df_gdp_literacy_merged['Literacy_Rate'].fillna(0) +
        df_gdp_literacy_merged['Avg_Years_Schooling'].fillna(0) * 10
    ) / 2
    # Replace 0s if both were NaN originally for a more accurate representation of 'NaN'
    df_gdp_literacy_merged.loc[df_gdp_literacy_merged['Literacy_Rate'].isna() & df_gdp_literacy_merged['Avg_Years_Schooling'].isna(), 'Education_Index'] = pd.NA

    # Recalculate correlation matrix using df_literacy_schooling (original one before cleaning other DFs)
    numeric_cols_schooling = df_literacy_schooling.select_dtypes(include=['number']).columns
    correlation_matrix = df_literacy_schooling[numeric_cols_schooling].corr()

    return df_gdp_literacy_merged, df_adult_youth_literacy_cleaned, correlation_matrix, df_literacy_schooling

# Load data (this will run only once due to caching)
df_gdp_literacy_merged, df_adult_youth_literacy_cleaned, correlation_matrix, df_literacy_schooling = load_and_prepare_all_data()

# Ensure altair.data_transformers is disabled for large datasets if needed
alt.data_transformers.disable_max_rows()

# --- 1. Education_Index vs. GDP_per_Capita Scatter Plot ---
st.header("1. Education Index vs. GDP per Capita")
df_plot_gdp_education = df_gdp_literacy_merged.dropna(subset=['Education_Index', 'GDP_per_Capita']).copy()
chart_gdp_education = alt.Chart(df_plot_gdp_education).mark_point().encode(
    x=alt.X('Education_Index:Q', title='Education Index'),
    y=alt.Y('GDP_per_Capita:Q', title='GDP per Capita'),
    color=alt.Color('entity:N', title='Country'),
    tooltip=[
        alt.Tooltip('entity:N', title='Country'),
        alt.Tooltip('year:Q', title='Year'),
        alt.Tooltip('Education_Index:Q', title='Education Index', format='.2f'),
        alt.Tooltip('GDP_per_Capita:Q', title='GDP per Capita', format='.2s')
    ]
).properties(
    title='Education Index vs. GDP per Capita'
).interactive()
st.altair_chart(chart_gdp_education, use_container_width=True)

# --- 2. Adult Literacy Rate vs. Average Years of Schooling Scatter Plot ---
st.header("2. Adult Literacy Rate vs. Average Years of Schooling")
df_plot_literacy_schooling = df_gdp_literacy_merged.dropna(subset=['Literacy_Rate', 'Avg_Years_Schooling']).copy()
chart_literacy_schooling = alt.Chart(df_plot_literacy_schooling).mark_point().encode(
    x=alt.X('Avg_Years_Schooling:Q', title='Average Years of Schooling'),
    y=alt.Y('Literacy_Rate:Q', title='Adult Literacy Rate (%)'),
    color=alt.Color('entity:N', title='Country'),
    tooltip=[
        alt.Tooltip('entity:N', title='Country'),
        alt.Tooltip('year:Q', title='Year'),
        alt.Tooltip('Literacy_Rate:Q', title='Adult Literacy Rate', format='.2f'),
        alt.Tooltip('Avg_Years_Schooling:Q', title='Avg. Years Schooling', format='.2f')
    ]
).properties(
    title='Adult Literacy Rate vs. Average Years of Schooling'
).interactive()
st.altair_chart(chart_literacy_schooling, use_container_width=True)

# --- 3. Literacy Gender Gap Over Time Line Chart ---
st.header("3. Literacy Gender Gap Over Time for Selected Countries")
selected_countries_gender_gap = ['India', 'China', 'Afghanistan', 'Albania', 'Brazil']
df_gender_gap_plot = df_adult_youth_literacy_cleaned[
    df_adult_youth_literacy_cleaned['entity'].isin(selected_countries_gender_gap)
].dropna(subset=['Literacy_Gender_Gap']).copy()
chart_gender_gap_time = alt.Chart(df_gender_gap_plot).mark_line(point=True).encode(
    x=alt.X('year:Q', axis=alt.Axis(format='d'), title='Year'),
    y=alt.Y('Literacy_Gender_Gap:Q', title='Literacy Gender Gap (%)'),
    color=alt.Color('entity:N', title='Country'),
    tooltip=[
        alt.Tooltip('entity:N', title='Country'),
        alt.Tooltip('year:Q', title='Year'),
        alt.Tooltip('Literacy_Gender_Gap:Q', title='Literacy Gender Gap', format='.2f')
    ]
).properties(
    title='Literacy Gender Gap Over Time for Selected Countries'
).interactive()
st.altair_chart(chart_gender_gap_time, use_container_width=True)

# --- 4. Histograms for Key Numerical Features ---
st.header("4. Distributions of Key Numerical Features (Histograms)")

numerical_features = [
    'GDP_per_Capita',
    'Avg_Years_Schooling',
    'Literacy_Rate',
    'Youth_Literacy_Average',
    'Education_Index'
]

for col in numerical_features:
    # Only plot if column exists and has non-null values
    if col in df_gdp_literacy_merged.columns and not df_gdp_literacy_merged[col].dropna().empty:
        fig, ax = plt.subplots(figsize=(8, 5))
        sns.histplot(df_gdp_literacy_merged[col].dropna(), kde=True, ax=ax)
        ax.set_title(f'Distribution of {col}')
        ax.set_xlabel(col)
        ax.set_ylabel('Frequency')
        ax.grid(True, linestyle='--', alpha=0.6)
        st.pyplot(fig)
        plt.close(fig)
    else:
        st.write(f"No data to display histogram for {col}.")

# --- 5. Correlation Heatmap ---
st.header("5. Correlation Heatmap of Numeric Columns")
if not correlation_matrix.empty:
    fig, ax = plt.subplots(figsize=(10, 8))
    sns.heatmap(correlation_matrix,
                annot=True,
                cmap='coolwarm',
                fmt='.2f',
                linewidths=.5,
                ax=ax)
    ax.set_title('Correlation Matrix of Numeric Columns')
    st.pyplot(fig)
    plt.close(fig)
else:
    st.write("No data to display correlation heatmap.")
````

````python
import streamlit as st
import pandas as pd
import altair as alt
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import io

st.set_page_config(
    page_title="EDA Visualizations",
    page_icon="📈",
)

st.title("Exploratory Data Analysis Visualizations")
st.write("View interactive charts and graphs for key educational and economic indicators.")

# User-Agent for data fetching
headers = {'User-Agent': 'Our World In Data data fetch/1.0'}

@st.cache_data
def load_and_prepare_all_data():
    """
    Loads, cleans, and feature-engineers all necessary dataframes.
    This function is cached to run only once.
    """
    # 1. Load initial datasets
    # Removed useColumnShortNames=true to ensure the expected long column name is present
    df_literacy = pd.read_csv("https://ourworldindata.org/grapher/literacy-rate-adults.csv?v=1&csvType=full", storage_options={'User-Agent': 'Our World In Data data fetch/1.0'})
    youth_literacy_url = "https://ourworldindata.org/grapher/literacy-rate-of-young-men-and-women.csv?v=1&csvType=full&useColumnShortNames=true"
    response_youth_literacy = requests.get(youth_literacy_url, headers=headers)
    response_youth_literacy.raise_for_status()
    df_youth_literacy = pd.read_csv(io.StringIO(response_youth_literacy.text))

    illiterate_pop_url = "https://ourworldindata.org/grapher/literate-and-illiterate-world-population.csv?v=1&csvType=full&useColumnShortNames=true"
    response_illiterate_pop = requests.get(illiterate_pop_url, headers=headers)
    response_illiterate_pop.raise_for_status()
    df_illiterate_pop = pd.read_csv(io.StringIO(response_illiterate_pop.text))

    gdp_per_capita_url = "https://ourworldindata.org/grapher/gdp-per-capita-worldbank.csv?v=1&csvType=full&useColumnShortNames=true"
    response_gdp = requests.get(gdp_per_capita_url, headers=headers)
    response_gdp.raise_for_status()
    df_gdp = pd.read_csv(io.StringIO(response_gdp.text))

    new_data_url = "https://ourworldindata.org/grapher/literacy-rates-vs-average-years-of-schooling.csv?v=1&csvType=full&useColumnShortNames=true"
    df_literacy_schooling = pd.read_csv(new_data_url, storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})

    # 2. Cleaning and Preprocessing (as done in notebook)

    # df_literacy cleaning
    df_literacy = df_literacy.rename(columns={'adult_literacy_rate__population_15plus_years__both_sexes__pct__lr_ag15t99': 'Literacy_Rate'})

    # df_youth_literacy cleaning
    df_youth_literacy_cleaned = df_youth_literacy.dropna(subset=['youth_literacy_rate__population_15_24_years__male__pct__lr_ag15t24_m', 'youth_literacy_rate__population_15_24_years__female__pct__lr_ag15t24_f']).copy()

    # Merging df_literacy and df_youth_literacy
    df_adult_youth_literacy = pd.merge(df_literacy, df_youth_literacy_cleaned, on=['entity', 'code', 'year'], how='outer', suffixes=('_adult', '_youth'))

    # df_adult_youth_literacy_cleaned (from cleaning stage)
    critical_literacy_cols = [
        'Literacy_Rate',
        'youth_literacy_rate__population_15_24_years__male__pct__lr_ag15t24_m',
        'youth_literacy_rate__population_15_24_years__female__pct__lr_ag15t24_f'
    ]
    df_adult_youth_literacy_cleaned = df_adult_youth_literacy.dropna(subset=critical_literacy_cols).copy()
    df_adult_youth_literacy_cleaned = df_adult_youth_literacy_cleaned.rename(columns={
        'youth_literacy_rate__population_15_24_years__male__pct__lr_ag15t24_m': 'Youth_Literacy_Male',
        'youth_literacy_rate__population_15_24_years__female__pct__lr_ag15t24_f': 'Youth_Literacy_Female'
    })
    df_adult_youth_literacy_cleaned = df_adult_youth_literacy_cleaned[
        (df_adult_youth_literacy_cleaned['year'] >= 1990) &
        (df_adult_youth_literacy_cleaned['year'] <= 2023)
    ].copy()

    # df_illiterate_pop_cleaned (from cleaning stage)
    df_illiterate_pop_cleaned = df_illiterate_pop.dropna(subset=['illiteracy_rate', 'literacy_rate']).copy()
    df_illiterate_pop_cleaned = df_illiterate_pop_cleaned[
        (df_illiterate_pop_cleaned['year'] >= 1990) &
        (df_illiterate_pop_cleaned['year'] <= 2023)
    ].copy()

    # df_gdp_schooling_cleaned (from cleaning stage)
    df_gdp_schooling = pd.merge(
        df_gdp, df_literacy_schooling,
        on=['entity', 'code', 'year'], how='outer',
        suffixes=('_gdp', '_schooling')
    )
    critical_gdp_schooling_cols = [
        'ny_gdp_pcap_pp_kd',
        'mf_youth_and_adults__15_64_years__average_years_of_education'
    ]
    df_gdp_schooling_cleaned = df_gdp_schooling.dropna(subset=critical_gdp_schooling_cols).copy()
    df_gdp_schooling_cleaned = df_gdp_schooling_cleaned.rename(columns={
        'ny_gdp_pcap_pp_kd': 'GDP_per_Capita',
        'mf_youth_and_adults__15_64_years__average_years_of_education': 'Avg_Years_Schooling',
        'population_historical': 'Population_Historical'
    })
    df_gdp_schooling_cleaned = df_gdp_schooling_cleaned[
        (df_gdp_schooling_cleaned['year'] >= 1990) &
        (df_gdp_schooling_cleaned['year'] <= 2023)
    ].copy()

    # 3. Feature Engineering (as done in notebook)
    df_illiterate_pop_cleaned['Illiteracy_Percentage'] = df_illiterate_pop_cleaned['illiteracy_rate']
    df_adult_youth_literacy_cleaned['Literacy_Gender_Gap'] = abs(df_adult_youth_literacy_cleaned['Youth_Literacy_Male'] - df_adult_youth_literacy_cleaned['Youth_Literacy_Female'])
    df_adult_youth_literacy_cleaned['Youth_Literacy_Average'] = (df_adult_youth_literacy_cleaned['Youth_Literacy_Male'] + df_adult_youth_literacy_cleaned['Youth_Literacy_Female']) / 2
    df_adult_youth_literacy_cleaned['Literacy_Growth_Rate'] = df_adult_youth_literacy_cleaned.groupby('entity')['Literacy_Rate'].pct_change() * 100

    # Merge df_gdp_schooling_cleaned and df_adult_youth_literacy_cleaned for combined features
    df_gdp_literacy_merged = pd.merge(
        df_gdp_schooling_cleaned,
        df_adult_youth_literacy_cleaned,
        on=['entity', 'code', 'year'],
        how='outer',
        suffixes=('_gdp_schooling', '_adult_youth')
    )

    df_gdp_literacy_merged['GDP_per_Schooling_Year'] = df_gdp_literacy_merged['GDP_per_Capita'] / df_gdp_literacy_merged['Avg_Years_Schooling']
    # Recalculate Youth_Literacy_Average on df_gdp_literacy_merged as it might have NaNs from outer join
    df_gdp_literacy_merged['Youth_Literacy_Average'] = (df_gdp_literacy_merged['Youth_Literacy_Male'] + df_gdp_literacy_merged['Youth_Literacy_Female']) / 2

    # Ensure Literacy_Rate and Avg_Years_Schooling are not NaN before calculating Education_Index
    df_gdp_literacy_merged['Education_Index'] = (
        df_gdp_literacy_merged['Literacy_Rate'].fillna(0) +
        df_gdp_literacy_merged['Avg_Years_Schooling'].fillna(0) * 10
    ) / 2
    # Replace 0s if both were NaN originally for a more accurate representation of 'NaN'
    df_gdp_literacy_merged.loc[df_gdp_literacy_merged['Literacy_Rate'].isna() & df_gdp_literacy_merged['Avg_Years_Schooling'].isna(), 'Education_Index'] = pd.NA

    # Recalculate correlation matrix using df_literacy_schooling (original one before cleaning other DFs)
    numeric_cols_schooling = df_literacy_schooling.select_dtypes(include=['number']).columns
    correlation_matrix = df_literacy_schooling[numeric_cols_schooling].corr()

    return df_gdp_literacy_merged, df_adult_youth_literacy_cleaned, correlation_matrix, df_literacy_schooling

# Load data (this will run only once due to caching)
df_gdp_literacy_merged, df_adult_youth_literacy_cleaned, correlation_matrix, df_literacy_schooling = load_and_prepare_all_data()

# Ensure altair.data_transformers is disabled for large datasets if needed
alt.data_transformers.disable_max_rows()

# --- 1. Education_Index vs. GDP_per_Capita Scatter Plot ---
st.header("1. Education Index vs. GDP per Capita")
df_plot_gdp_education = df_gdp_literacy_merged.dropna(subset=['Education_Index', 'GDP_per_Capita']).copy()
chart_gdp_education = alt.Chart(df_plot_gdp_education).mark_point().encode(
    x=alt.X('Education_Index:Q', title='Education Index'),
    y=alt.Y('GDP_per_Capita:Q', title='GDP per Capita'),
    color=alt.Color('entity:N', title='Country'),
    tooltip=[
        alt.Tooltip('entity:N', title='Country'),
        alt.Tooltip('year:Q', title='Year'),
        alt.Tooltip('Education_Index:Q', title='Education Index', format='.2f'),
        alt.Tooltip('GDP_per_Capita:Q', title='GDP per Capita', format='.2s')
    ]
).properties(
    title='Education Index vs. GDP per Capita'
).interactive()
st.altair_chart(chart_gdp_education, use_container_width=True)

# --- 2. Adult Literacy Rate vs. Average Years of Schooling Scatter Plot ---
st.header("2. Adult Literacy Rate vs. Average Years of Schooling")
df_plot_literacy_schooling = df_gdp_literacy_merged.dropna(subset=['Literacy_Rate', 'Avg_Years_Schooling']).copy()
chart_literacy_schooling = alt.Chart(df_plot_literacy_schooling).mark_point().encode(
    x=alt.X('Avg_Years_Schooling:Q', title='Average Years of Schooling'),
    y=alt.Y('Literacy_Rate:Q', title='Adult Literacy Rate (%)'),
    color=alt.Color('entity:N', title='Country'),
    tooltip=[
        alt.Tooltip('entity:N', title='Country'),
        alt.Tooltip('year:Q', title='Year'),
        alt.Tooltip('Literacy_Rate:Q', title='Adult Literacy Rate', format='.2f'),
        alt.Tooltip('Avg_Years_Schooling:Q', title='Avg. Years Schooling', format='.2f')
    ]
).properties(
    title='Adult Literacy Rate vs. Average Years of Schooling'
).interactive()
st.altair_chart(chart_literacy_schooling, use_container_width=True)

# --- 3. Literacy Gender Gap Over Time Line Chart ---
st.header("3. Literacy Gender Gap Over Time for Selected Countries")
selected_countries_gender_gap = ['India', 'China', 'Afghanistan', 'Albania', 'Brazil']
df_gender_gap_plot = df_adult_youth_literacy_cleaned[
    df_adult_youth_literacy_cleaned['entity'].isin(selected_countries_gender_gap)
].dropna(subset=['Literacy_Gender_Gap']).copy()
chart_gender_gap_time = alt.Chart(df_gender_gap_plot).mark_line(point=True).encode(
    x=alt.X('year:Q', axis=alt.Axis(format='d'), title='Year'),
    y=alt.Y('Literacy_Gender_Gap:Q', title='Literacy Gender Gap (%)'),
    color=alt.Color('entity:N', title='Country'),
    tooltip=[
        alt.Tooltip('entity:N', title='Country'),
        alt.Tooltip('year:Q', title='Year'),
        alt.Tooltip('Literacy_Gender_Gap:Q', title='Literacy Gender Gap', format='.2f')
    ]
).properties(
    title='Literacy Gender Gap Over Time for Selected Countries'
).interactive()
st.altair_chart(chart_gender_gap_time, use_container_width=True)

# --- 4. Histograms for Key Numerical Features ---
st.header("4. Distributions of Key Numerical Features (Histograms)")

numerical_features = [
    'GDP_per_Capita',
    'Avg_Years_Schooling',
    'Literacy_Rate',
    'Youth_Literacy_Average',
    'Education_Index'
]

for col in numerical_features:
    # Only plot if column exists and has non-null values
    if col in df_gdp_literacy_merged.columns and not df_gdp_literacy_merged[col].dropna().empty:
        fig, ax = plt.subplots(figsize=(8, 5))
        sns.histplot(df_gdp_literacy_merged[col].dropna(), kde=True, ax=ax)
        ax.set_title(f'Distribution of {col}')
        ax.set_xlabel(col)
        ax.set_ylabel('Frequency')
        ax.grid(True, linestyle='--', alpha=0.6)
        st.pyplot(fig)
        plt.close(fig)
    else:
        st.write(f"No data to display histogram for {col}.")

# --- 5. Correlation Heatmap ---
st.header("5. Correlation Heatmap of Numeric Columns")
if not correlation_matrix.empty:
    fig, ax = plt.subplots(figsize=(10, 8))
    sns.heatmap(correlation_matrix,
                annot=True,
                cmap='coolwarm',
                fmt='.2f',
                linewidths=.5,
                ax=ax)
    ax.set_title('Correlation Matrix of Numeric Columns')
    st.pyplot(fig)
    plt.close(fig)
else:
    st.write("No data to display correlation heatmap.")
````

````python
import streamlit as st
import pandas as pd
import altair as alt
import requests
import io

st.set_page_config(
    page_title="Country Profile",
    page_icon="🌍",
)

st.title("Country Profile")
st.write("Dive deep into the literacy and education statistics of specific countries over time.")

# User-Agent for data fetching
headers = {'User-Agent': 'Our World In Data data fetch/1.0'}

@st.cache_data
def load_and_prepare_all_data():
    """
    Loads, cleans, and feature-engineers all necessary dataframes.
    This function is cached to run only once.
    """
    # 1. Load initial datasets with consistent short column names
    df_literacy = pd.read_csv("https://ourworldindata.org/grapher/literacy-rate-adults.csv?v=1&csvType=full&useColumnShortNames=true", storage_options=headers)
    df_literacy.rename(columns={'Entity': 'entity', 'Code': 'code'}, inplace=True)

    youth_literacy_url = "https://ourworldindata.org/grapher/literacy-rate-of-young-men-and-women.csv?v=1&csvType=full&useColumnShortNames=true"
    response_youth_literacy = requests.get(youth_literacy_url, headers=headers)
    response_youth_literacy.raise_for_status()
    df_youth_literacy = pd.read_csv(io.StringIO(response_youth_literacy.text))
    df_youth_literacy.rename(columns={'Entity': 'entity', 'Code': 'code'}, inplace=True)

    illiterate_pop_url = "https://ourworldindata.org/grapher/literate-and-illiterate-world-population.csv?v=1&csvType=full&useColumnShortNames=true"
    response_illiterate_pop = requests.get(illiterate_pop_url, headers=headers)
    response_illiterate_pop.raise_for_status()
    df_illiterate_pop = pd.read_csv(io.StringIO(response_illiterate_pop.text))
    df_illiterate_pop.rename(columns={'Entity': 'entity', 'Code': 'code'}, inplace=True)

    gdp_per_capita_url = "https://ourworldindata.org/grapher/gdp-per-capita-worldbank.csv?v=1&csvType=full&useColumnShortNames=true"
    response_gdp = requests.get(gdp_per_capita_url, headers=headers)
    response_gdp.raise_for_status()
    df_gdp = pd.read_csv(io.StringIO(response_gdp.text))
    df_gdp.rename(columns={'Entity': 'entity', 'Code': 'code'}, inplace=True)

    new_data_url = "https://ourworldindata.org/grapher/literacy-rates-vs-average-years-of-schooling.csv?v=1&csvType=full&useColumnShortNames=true"
    df_literacy_schooling = pd.read_csv(new_data_url, storage_options=headers)
    df_literacy_schooling.rename(columns={'Entity': 'entity', 'Code': 'code'}, inplace=True)

    # 2. Cleaning and Preprocessing (as done in notebook)

    # df_literacy cleaning
    # Column name is already short form due to useColumnShortNames=true
    df_literacy = df_literacy.rename(columns={'adult_literacy_rate__population_15plus_years__both_sexes__pct__lr_ag15t99': 'Literacy_Rate'})

    # df_youth_literacy cleaning - use short column names for dropna and rename
    df_youth_literacy_cleaned = df_youth_literacy.dropna(subset=[
        'youth_literacy_rate__population_15_24_years__male__pct__lr_ag15t24_m',
        'youth_literacy_rate__population_15_24_years__female__pct__lr_ag15t24_f'
    ]).copy()

    # Merging df_literacy and df_youth_literacy
    df_adult_youth_literacy = pd.merge(df_literacy, df_youth_literacy_cleaned, on=['entity', 'code', 'year'], how='outer', suffixes=('_adult', '_youth'))

    # df_adult_youth_literacy_cleaned (from cleaning stage)
    critical_literacy_cols = [
        'Literacy_Rate',
        'youth_literacy_rate__population_15_24_years__male__pct__lr_ag15t24_m',
        'youth_literacy_rate__population_15_24_years__female__pct__lr_ag15t24_f'
    ]
    df_adult_youth_literacy_cleaned = df_adult_youth_literacy.dropna(subset=critical_literacy_cols).copy()
    df_adult_youth_literacy_cleaned = df_adult_youth_literacy_cleaned.rename(columns={
        'youth_literacy_rate__population_15_24_years__male__pct__lr_ag15t24_m': 'Youth_Literacy_Male',
        'youth_literacy_rate__population_15_24_years__female__pct__lr_ag15t24_f': 'Youth_Literacy_Female'
    })
    df_adult_youth_literacy_cleaned = df_adult_youth_literacy_cleaned[
        (df_adult_youth_literacy_cleaned['year'] >= 1990) &
        (df_adult_youth_literacy_cleaned['year'] <= 2023)
    ].copy()

    # df_illiterate_pop_cleaned (from cleaning stage)
    # Column names are already short form due to useColumnShortNames=true or naturally short
    df_illiterate_pop_cleaned = df_illiterate_pop.dropna(subset=['illiteracy_rate', 'literacy_rate']).copy()
    df_illiterate_pop_cleaned = df_illiterate_pop_cleaned[
        (df_illiterate_pop_cleaned['year'] >= 1990) &
        (df_illiterate_pop_cleaned['year'] <= 2023)
    ].copy()

    # df_gdp_schooling_cleaned (from cleaning stage) - use short column names
    df_gdp_schooling = pd.merge(
        df_gdp, df_literacy_schooling,
        on=['entity', 'code', 'year'], how='outer',
        suffixes=('_gdp', '_schooling')
    )
    critical_gdp_schooling_cols = [
        'ny_gdp_pcap_pp_kd', # short name from df_gdp
        'mf_youth_and_adults__15_64_years__average_years_of_education' # short name from df_literacy_schooling
    ]
    df_gdp_schooling_cleaned = df_gdp_schooling.dropna(subset=critical_gdp_schooling_cols).copy()
    df_gdp_schooling_cleaned = df_gdp_schooling_cleaned.rename(columns={
        'ny_gdp_pcap_pp_kd': 'GDP_per_Capita',
        'mf_youth_and_adults__15_64_years__average_years_of_education': 'Avg_Years_Schooling',
        'population_historical': 'Population_Historical' # short name from df_literacy_schooling
    })
    df_gdp_schooling_cleaned = df_gdp_schooling_cleaned[
        (df_gdp_schooling_cleaned['year'] >= 1990) &
        (df_gdp_schooling_cleaned['year'] <= 2023)
    ].copy()

    # 3. Feature Engineering (as done in notebook)
    df_illiterate_pop_cleaned['Illiteracy_Percentage'] = df_illiterate_pop_cleaned['illiteracy_rate']
    df_adult_youth_literacy_cleaned['Literacy_Gender_Gap'] = abs(df_adult_youth_literacy_cleaned['Youth_Literacy_Male'] - df_adult_youth_literacy_cleaned['Youth_Literacy_Female'])
    df_adult_youth_literacy_cleaned['Youth_Literacy_Average'] = (df_adult_youth_literacy_cleaned['Youth_Literacy_Male'] + df_adult_youth_literacy_cleaned['Youth_Literacy_Female']) / 2
    df_adult_youth_literacy_cleaned['Literacy_Growth_Rate'] = df_adult_youth_literacy_cleaned.groupby('entity')['Literacy_Rate'].pct_change() * 100

    # Merge df_gdp_schooling_cleaned and df_adult_youth_literacy_cleaned for combined features
    df_gdp_literacy_merged = pd.merge(
        df_gdp_schooling_cleaned,
        df_adult_youth_literacy_cleaned,
        on=['entity', 'code', 'year'],
        how='outer',
        suffixes=('_gdp_schooling', '_adult_youth')
    )

    df_gdp_literacy_merged['GDP_per_Schooling_Year'] = df_gdp_literacy_merged['GDP_per_Capita'] / df_gdp_literacy_merged['Avg_Years_Schooling']
    df_gdp_literacy_merged['Youth_Literacy_Average'] = (df_gdp_literacy_merged['Youth_Literacy_Male'] + df_gdp_literacy_merged['Youth_Literacy_Female']) / 2

    df_gdp_literacy_merged['Education_Index'] = (
        df_gdp_literacy_merged['Literacy_Rate'].fillna(0) +
        df_gdp_literacy_merged['Avg_Years_Schooling'].fillna(0) * 10
    ) / 2
    df_gdp_literacy_merged.loc[df_gdp_literacy_merged['Literacy_Rate'].isna() & df_gdp_literacy_merged['Avg_Years_Schooling'].isna(), 'Education_Index'] = pd.NA

    return df_gdp_literacy_merged, df_adult_youth_literacy_cleaned

# Load data (this will run only once due to caching)
df_gdp_literacy_merged, df_adult_youth_literacy_cleaned = load_and_prepare_all_data()

alt.data_transformers.disable_max_rows()

# Country Selection
available_countries = sorted(df_gdp_literacy_merged['entity'].dropna().unique().tolist())
selected_country = st.sidebar.selectbox(
    "Select a Country",
    available_countries
)

if selected_country:
    st.header(f"Profile for {selected_country}")

    # Filter data for the selected country
    country_gdp_literacy = df_gdp_literacy_merged[df_gdp_literacy_merged['entity'] == selected_country].copy()
    country_adult_youth_literacy = df_adult_youth_literacy_cleaned[df_adult_youth_literacy_cleaned['entity'] == selected_country].copy()

    if not country_gdp_literacy.empty or not country_adult_youth_literacy.empty:
        # 1. Adult Literacy Rate Over Time
        if not country_gdp_literacy.dropna(subset=['Literacy_Rate']).empty:
            chart_adult_literacy = alt.Chart(country_gdp_literacy).mark_line(point=True).encode(
                x=alt.X('year:Q', axis=alt.Axis(format='d'), title='Year'),
                y=alt.Y('Literacy_Rate:Q', title='Adult Literacy Rate (%)'),
                tooltip=['entity:N', 'year:Q', alt.Tooltip('Literacy_Rate:Q', format='.2f')]
            ).properties(
                title=f'{selected_country}: Adult Literacy Rate Over Time'
            ).interactive()
            st.altair_chart(chart_adult_literacy, use_container_width=True)
        else:
            st.write(f"No adult literacy rate data available for {selected_country}.")

        # 2. Youth Literacy Rates Over Time
        if not country_adult_youth_literacy.dropna(subset=['Youth_Literacy_Male', 'Youth_Literacy_Female', 'Youth_Literacy_Average']).empty:
            df_youth_melted = country_adult_youth_literacy.melt(
                id_vars=['entity', 'year'],
                value_vars=['Youth_Literacy_Male', 'Youth_Literacy_Female', 'Youth_Literacy_Average'],
                var_name='Gender_Metric',
                value_name='Youth_Literacy_Value' # Changed value_name here
            )
            chart_youth_literacy = alt.Chart(df_youth_melted).mark_line(point=True).encode(
                x=alt.X('year:Q', axis=alt.Axis(format='d'), title='Year'),
                y=alt.Y('Youth_Literacy_Value:Q', title='Youth Literacy Rate (%)'), # Use the new value_name
                color=alt.Color('Gender_Metric:N', title='Metric'),
                tooltip=['entity:N', 'year:Q', 'Gender_Metric:N', alt.Tooltip('Youth_Literacy_Value:Q', format='.2f')] # Use the new value_name
            ).properties(
                title=f'{selected_country}: Youth Literacy Rates Over Time'
            ).interactive()
            st.altair_chart(chart_youth_literacy, use_container_width=True)
        else:
            st.write(f"No youth literacy rate data available for {selected_country}.")

        # 3. Literacy Gender Gap Over Time
        if not country_adult_youth_literacy.dropna(subset=['Literacy_Gender_Gap']).empty:
            chart_gender_gap = alt.Chart(country_adult_youth_literacy).mark_line(point=True).encode(
                x=alt.X('year:Q', axis=alt.Axis(format='d'), title='Year'),
                y=alt.Y('Literacy_Gender_Gap:Q', title='Literacy Gender Gap (%)'),
                tooltip=['entity:N', 'year:Q', alt.Tooltip('Literacy_Gender_Gap:Q', format='.2f')]
            ).properties(
                title=f'{selected_country}: Literacy Gender Gap Over Time'
            ).interactive()
            st.altair_chart(chart_gender_gap, use_container_width=True)
        else:
            st.write(f"No literacy gender gap data available for {selected_country}.")

        # 4. GDP per Capita Over Time
        if not country_gdp_literacy.dropna(subset=['GDP_per_Capita']).empty:
            chart_gdp = alt.Chart(country_gdp_literacy).mark_line(point=True).encode(
                x=alt.X('year:Q', axis=alt.Axis(format='d'), title='Year'),
                y=alt.Y('GDP_per_Capita:Q', title='GDP per Capita'),
                tooltip=['entity:N', 'year:Q', alt.Tooltip('GDP_per_Capita:Q', format='.2s')]
            ).properties(
                title=f'{selected_country}: GDP per Capita Over Time'
            ).interactive()
            st.altair_chart(chart_gdp, use_container_width=True)
        else:
            st.write(f"No GDP per capita data available for {selected_country}.")

        # 5. Average Years of Schooling Over Time
        if not country_gdp_literacy.dropna(subset=['Avg_Years_Schooling']).empty:
            chart_schooling = alt.Chart(country_gdp_literacy).mark_line(point=True).encode(
                x=alt.X('year:Q', axis=alt.Axis(format='d'), title='Year'),
                y=alt.Y('Avg_Years_Schooling:Q', title='Average Years of Schooling'),
                tooltip=['entity:N', 'year:Q', alt.Tooltip('Avg_Years_Schooling:Q', format='.2f')]
            ).properties(
                title=f'{selected_country}: Average Years of Schooling Over Time'
            ).interactive()
            st.altair_chart(chart_schooling, use_container_width=True)
        else:
            st.write(f"No average years of schooling data available for {selected_country}.")

        # 6. Education Index Over Time
        if not country_gdp_literacy.dropna(subset=['Education_Index']).empty:
            chart_education_index = alt.Chart(country_gdp_literacy).mark_line(point=True).encode(
                x=alt.X('year:Q', axis=alt.Axis(format='d'), title='Year'),
                y=alt.Y('Education_Index:Q', title='Education Index'),
                tooltip=['entity:N', 'year:Q', alt.Tooltip('Education_Index:Q', format='.2f')]
            ).properties(
                title=f'{selected_country}: Education Index Over Time'
            ).interactive()
            st.altair_chart(chart_education_index, use_container_width=True)
        else:
            st.write(f"No Education Index data available for {selected_country}.")
    else:
        st.write(f"No data available for {selected_country} across any indicators.")
else:
    st.info("Please select a country from the sidebar to view its profile.")
````

**Reasoning**:
The `ValueError` occurred because the `value_name` parameter in the `melt` function (`'Literacy_Rate'`) already existed as a column in the `country_adult_youth_literacy` DataFrame. To resolve this, I will change `value_name` to `'Youth_Literacy_Value'` in the `melt` operation to avoid the column name clash and update the chart encoding accordingly.



````python
import streamlit as st
import pandas as pd
import altair as alt
import requests
import io

st.set_page_config(
    page_title="Country Profile",
    page_icon="🌍",
)

st.title("Country Profile")
st.write("Dive deep into the literacy and education statistics of specific countries over time.")

# User-Agent for data fetching
headers = {'User-Agent': 'Our World In Data data fetch/1.0'}

@st.cache_data
def load_and_prepare_all_data():
    """
    Loads, cleans, and feature-engineers all necessary dataframes.
    This function is cached to run only once.
    """
    # 1. Load initial datasets with consistent short column names
    df_literacy = pd.read_csv("https://ourworldindata.org/grapher/literacy-rate-adults.csv?v=1&csvType=full&useColumnShortNames=true", storage_options=headers)
    df_literacy.rename(columns={'Entity': 'entity', 'Code': 'code'}, inplace=True)

    youth_literacy_url = "https://ourworldindata.org/grapher/literacy-rate-of-young-men-and-women.csv?v=1&csvType=full&useColumnShortNames=true"
    response_youth_literacy = requests.get(youth_literacy_url, headers=headers)
    response_youth_literacy.raise_for_status()
    df_youth_literacy = pd.read_csv(io.StringIO(response_youth_literacy.text))
    df_youth_literacy.rename(columns={'Entity': 'entity', 'Code': 'code'}, inplace=True)

    illiterate_pop_url = "https://ourworldindata.org/grapher/literate-and-illiterate-world-population.csv?v=1&csvType=full&useColumnShortNames=true"
    response_illiterate_pop = requests.get(illiterate_pop_url, headers=headers)
    response_illiterate_pop.raise_for_status()
    df_illiterate_pop = pd.read_csv(io.StringIO(response_illiterate_pop.text))
    df_illiterate_pop.rename(columns={'Entity': 'entity', 'Code': 'code'}, inplace=True)

    gdp_per_capita_url = "https://ourworldindata.org/grapher/gdp-per-capita-worldbank.csv?v=1&csvType=full&useColumnShortNames=true"
    response_gdp = requests.get(gdp_per_capita_url, headers=headers)
    response_gdp.raise_for_status()
    df_gdp = pd.read_csv(io.StringIO(response_gdp.text))
    df_gdp.rename(columns={'Entity': 'entity', 'Code': 'code'}, inplace=True)

    new_data_url = "https://ourworldindata.org/grapher/literacy-rates-vs-average-years-of-schooling.csv?v=1&csvType=full&useColumnShortNames=true"
    df_literacy_schooling = pd.read_csv(new_data_url, storage_options=headers)
    df_literacy_schooling.rename(columns={'Entity': 'entity', 'Code': 'code'}, inplace=True)

    # 2. Cleaning and Preprocessing (as done in notebook)

    # df_literacy cleaning
    # Column name is already short form due to useColumnShortNames=true
    df_literacy = df_literacy.rename(columns={'adult_literacy_rate__population_15plus_years__both_sexes__pct__lr_ag15t99': 'Literacy_Rate'})

    # df_youth_literacy cleaning - use short column names for dropna and rename
    df_youth_literacy_cleaned = df_youth_literacy.dropna(subset=[
        'youth_literacy_rate__population_15_24_years__male__pct__lr_ag15t24_m',
        'youth_literacy_rate__population_15_24_years__female__pct__lr_ag15t24_f'
    ]).copy()

    # Merging df_literacy and df_youth_literacy
    df_adult_youth_literacy = pd.merge(df_literacy, df_youth_literacy_cleaned, on=['entity', 'code', 'year'], how='outer', suffixes=('_adult', '_youth'))

    # df_adult_youth_literacy_cleaned (from cleaning stage)
    critical_literacy_cols = [
        'Literacy_Rate',
        'youth_literacy_rate__population_15_24_years__male__pct__lr_ag15t24_m',
        'youth_literacy_rate__population_15_24_years__female__pct__lr_ag15t24_f'
    ]
    df_adult_youth_literacy_cleaned = df_adult_youth_literacy.dropna(subset=critical_literacy_cols).copy()
    df_adult_youth_literacy_cleaned = df_adult_youth_literacy_cleaned.rename(columns={
        'youth_literacy_rate__population_15_24_years__male__pct__lr_ag15t24_m': 'Youth_Literacy_Male',
        'youth_literacy_rate__population_15_24_years__female__pct__lr_ag15t24_f': 'Youth_Literacy_Female'
    })
    df_adult_youth_literacy_cleaned = df_adult_youth_literacy_cleaned[
        (df_adult_youth_literacy_cleaned['year'] >= 1990) &
        (df_adult_youth_literacy_cleaned['year'] <= 2023)
    ].copy()

    # df_illiterate_pop_cleaned (from cleaning stage)
    # Column names are already short form due to useColumnShortNames=true or naturally short
    df_illiterate_pop_cleaned = df_illiterate_pop.dropna(subset=['illiteracy_rate', 'literacy_rate']).copy()
    df_illiterate_pop_cleaned = df_illiterate_pop_cleaned[
        (df_illiterate_pop_cleaned['year'] >= 1990) &
        (df_illiterate_pop_cleaned['year'] <= 2023)
    ].copy()

    # df_gdp_schooling_cleaned (from cleaning stage) - use short column names
    df_gdp_schooling = pd.merge(
        df_gdp, df_literacy_schooling,
        on=['entity', 'code', 'year'], how='outer',
        suffixes=('_gdp', '_schooling')
    )
    critical_gdp_schooling_cols = [
        'ny_gdp_pcap_pp_kd', # short name from df_gdp
        'mf_youth_and_adults__15_64_years__average_years_of_education' # short name from df_literacy_schooling
    ]
    df_gdp_schooling_cleaned = df_gdp_schooling.dropna(subset=critical_gdp_schooling_cols).copy()
    df_gdp_schooling_cleaned = df_gdp_schooling_cleaned.rename(columns={
        'ny_gdp_pcap_pp_kd': 'GDP_per_Capita',
        'mf_youth_and_adults__15_64_years__average_years_of_education': 'Avg_Years_Schooling',
        'population_historical': 'Population_Historical' # short name from df_literacy_schooling
    })
    df_gdp_schooling_cleaned = df_gdp_schooling_cleaned[
        (df_gdp_schooling_cleaned['year'] >= 1990) &
        (df_gdp_schooling_cleaned['year'] <= 2023)
    ].copy()

    # 3. Feature Engineering (as done in notebook)
    df_illiterate_pop_cleaned['Illiteracy_Percentage'] = df_illiterate_pop_cleaned['illiteracy_rate']
    df_adult_youth_literacy_cleaned['Literacy_Gender_Gap'] = abs(df_adult_youth_literacy_cleaned['Youth_Literacy_Male'] - df_adult_youth_literacy_cleaned['Youth_Literacy_Female'])
    df_adult_youth_literacy_cleaned['Youth_Literacy_Average'] = (df_adult_youth_literacy_cleaned['Youth_Literacy_Male'] + df_adult_youth_literacy_cleaned['Youth_Literacy_Female']) / 2
    df_adult_youth_literacy_cleaned['Literacy_Growth_Rate'] = df_adult_youth_literacy_cleaned.groupby('entity')['Literacy_Rate'].pct_change() * 100

    # Merge df_gdp_schooling_cleaned and df_adult_youth_literacy_cleaned for combined features
    df_gdp_literacy_merged = pd.merge(
        df_gdp_schooling_cleaned,
        df_adult_youth_literacy_cleaned,
        on=['entity', 'code', 'year'],
        how='outer',
        suffixes=('_gdp_schooling', '_adult_youth')
    )

    df_gdp_literacy_merged['GDP_per_Schooling_Year'] = df_gdp_literacy_merged['GDP_per_Capita'] / df_gdp_literacy_merged['Avg_Years_Schooling']
    df_gdp_literacy_merged['Youth_Literacy_Average'] = (df_gdp_literacy_merged['Youth_Literacy_Male'] + df_gdp_literacy_merged['Youth_Literacy_Female']) / 2

    df_gdp_literacy_merged['Education_Index'] = (
        df_gdp_literacy_merged['Literacy_Rate'].fillna(0) +
        df_gdp_literacy_merged['Avg_Years_Schooling'].fillna(0) * 10
    ) / 2
    df_gdp_literacy_merged.loc[df_gdp_literacy_merged['Literacy_Rate'].isna() & df_gdp_literacy_merged['Avg_Years_Schooling'].isna(), 'Education_Index'] = pd.NA

    return df_gdp_literacy_merged, df_adult_youth_literacy_cleaned

# Load data (this will run only once due to caching)
df_gdp_literacy_merged, df_adult_youth_literacy_cleaned = load_and_prepare_all_data()

alt.data_transformers.disable_max_rows()

# Country Selection
available_countries = sorted(df_gdp_literacy_merged['entity'].dropna().unique().tolist())
selected_country = st.sidebar.selectbox(
    "Select a Country",
    available_countries
)

if selected_country:
    st.header(f"Profile for {selected_country}")

    # Filter data for the selected country
    country_gdp_literacy = df_gdp_literacy_merged[df_gdp_literacy_merged['entity'] == selected_country].copy()
    country_adult_youth_literacy = df_adult_youth_literacy_cleaned[df_adult_youth_literacy_cleaned['entity'] == selected_country].copy()

    if not country_gdp_literacy.empty or not country_adult_youth_literacy.empty:
        # 1. Adult Literacy Rate Over Time
        if not country_gdp_literacy.dropna(subset=['Literacy_Rate']).empty:
            chart_adult_literacy = alt.Chart(country_gdp_literacy).mark_line(point=True).encode(
                x=alt.X('year:Q', axis=alt.Axis(format='d'), title='Year'),
                y=alt.Y('Literacy_Rate:Q', title='Adult Literacy Rate (%)'),
                tooltip=['entity:N', 'year:Q', alt.Tooltip('Literacy_Rate:Q', format='.2f')]
            ).properties(
                title=f'{selected_country}: Adult Literacy Rate Over Time'
            ).interactive()
            st.altair_chart(chart_adult_literacy, use_container_width=True)
        else:
            st.write(f"No adult literacy rate data available for {selected_country}.")

        # 2. Youth Literacy Rates Over Time
        if not country_adult_youth_literacy.dropna(subset=['Youth_Literacy_Male', 'Youth_Literacy_Female', 'Youth_Literacy_Average']).empty:
            df_youth_melted = country_adult_youth_literacy.melt(
                id_vars=['entity', 'year'],
                value_vars=['Youth_Literacy_Male', 'Youth_Literacy_Female', 'Youth_Literacy_Average'],
                var_name='Gender_Metric',
                value_name='Youth_Literacy_Value' # Changed value_name here
            )
            chart_youth_literacy = alt.Chart(df_youth_melted).mark_line(point=True).encode(
                x=alt.X('year:Q', axis=alt.Axis(format='d'), title='Year'),
                y=alt.Y('Youth_Literacy_Value:Q', title='Youth Literacy Rate (%)'), # Use the new value_name
                color=alt.Color('Gender_Metric:N', title='Metric'),
                tooltip=['entity:N', 'year:Q', 'Gender_Metric:N', alt.Tooltip('Youth_Literacy_Value:Q', format='.2f')] # Use the new value_name
            ).properties(
                title=f'{selected_country}: Youth Literacy Rates Over Time'
            ).interactive()
            st.altair_chart(chart_youth_literacy, use_container_width=True)
        else:
            st.write(f"No youth literacy rate data available for {selected_country}.")

        # 3. Literacy Gender Gap Over Time
        if not country_adult_youth_literacy.dropna(subset=['Literacy_Gender_Gap']).empty:
            chart_gender_gap = alt.Chart(country_adult_youth_literacy).mark_line(point=True).encode(
                x=alt.X('year:Q', axis=alt.Axis(format='d'), title='Year'),
                y=alt.Y('Literacy_Gender_Gap:Q', title='Literacy Gender Gap (%)'),
                tooltip=['entity:N', 'year:Q', alt.Tooltip('Literacy_Gender_Gap:Q', format='.2f')]
            ).properties(
                title=f'{selected_country}: Literacy Gender Gap Over Time'
            ).interactive()
            st.altair_chart(chart_gender_gap, use_container_width=True)
        else:
            st.write(f"No literacy gender gap data available for {selected_country}.")

        # 4. GDP per Capita Over Time
        if not country_gdp_literacy.dropna(subset=['GDP_per_Capita']).empty:
            chart_gdp = alt.Chart(country_gdp_literacy).mark_line(point=True).encode(
                x=alt.X('year:Q', axis=alt.Axis(format='d'), title='Year'),
                y=alt.Y('GDP_per_Capita:Q', title='GDP per Capita'),
                tooltip=['entity:N', 'year:Q', alt.Tooltip('GDP_per_Capita:Q', format='.2s')]
            ).properties(
                title=f'{selected_country}: GDP per Capita Over Time'
            ).interactive()
            st.altair_chart(chart_gdp, use_container_width=True)
        else:
            st.write(f"No GDP per capita data available for {selected_country}.")

        # 5. Average Years of Schooling Over Time
        if not country_gdp_literacy.dropna(subset=['Avg_Years_Schooling']).empty:
            chart_schooling = alt.Chart(country_gdp_literacy).mark_line(point=True).encode(
                x=alt.X('year:Q', axis=alt.Axis(format='d'), title='Year'),
                y=alt.Y('Avg_Years_Schooling:Q', title='Average Years of Schooling'),
                tooltip=['entity:N', 'year:Q', alt.Tooltip('Avg_Years_Schooling:Q', format='.2f')]
            ).properties(
                title=f'{selected_country}: Average Years of Schooling Over Time'
            ).interactive()
            st.altair_chart(chart_schooling, use_container_width=True)
        else:
            st.write(f"No average years of schooling data available for {selected_country}.")

        # 6. Education Index Over Time
        if not country_gdp_literacy.dropna(subset=['Education_Index']).empty:
            chart_education_index = alt.Chart(country_gdp_literacy).mark_line(point=True).encode(
                x=alt.X('year:Q', axis=alt.Axis(format='d'), title='Year'),
                y=alt.Y('Education_Index:Q', title='Education Index'),
                tooltip=['entity:N', 'year:Q', alt.Tooltip('Education_Index:Q', format='.2f')]
            ).properties(
                title=f'{selected_country}: Education Index Over Time'
            ).interactive()
            st.altair_chart(chart_education_index, use_container_width=True)
        else:
            st.write(f"No Education Index data available for {selected_country}.")
    else:
        st.write(f"No data available for {selected_country} across any indicators.")
else:
    st.info("Please select a country from the sidebar to view its profile.")
````
